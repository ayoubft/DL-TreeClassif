```python
import geopandas as gpd
import pandas as pd
import os
import shutil
```

```python
# !rm -rf ./data/0 ./data/1 ./data/2 ./data/3 ./data/4 ./data/5 ./data/6
```

```python
file = "./data/BBtrees.dbf"
Table = gpd.read_file(file)
gtable = pd.DataFrame(Table)
```

```python
gtable
```

```python
targets = list(gtable["target"].unique())
```

```python
targets
```

```python
d = {}
for t in targets:
    d[t] = list(gtable[gtable["target"] == t]["id"])
```

```python
print(list(d.keys()))
```

```python
root_path = "/home/ayoub/Projects/MINI-PROJET-teled/DL-tree3BinaryClass"

for item in list(d.keys()):
    os.mkdir("./data/" + str(item))
```

```python
d2 = {}

for k, v in d.items():
    l_int = []
    for id_ in v:
        l_int.append("./data/indivTreesJPG/tid_" + str(id_) + ".jpg")
    d2[k] = l_int
```

```python
for k, v in d2.items():
    files = v
    for f in files:
        shutil.copy(f, f"./data/{k}/")
```
